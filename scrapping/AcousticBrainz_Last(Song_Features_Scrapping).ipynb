```python
#Last Tapi Lengkap
import requests
import time
import json
import os

# Daftar MBID dari hasil pencarian MusicBrainz
# mbid_list = [
#     "a0db51d6-9ade-49b1-8df7-f1959aa56d74",
# ]


def clean_mbid(mbid):
    """ Membersihkan MBID dari spasi atau karakter tersembunyi lainnya """
    return str(mbid).strip()


# File cache untuk menyimpan hasil query sebelumnya agar tidak perlu request ulang
cache_file = "acousticbrainz_cache.json"

# Load cache jika ada
if os.path.exists(cache_file):
    with open(cache_file, "r") as f:
        cache = json.load(f)
else:
    cache = {}

# Fungsi untuk mengambil data dari AcousticBrainz dengan handling rate limit
def get_acousticbrainz_data(mbid):
    if mbid in cache:
        print(f"Using cached data for MBID: {mbid}")
        return cache[mbid]["low_level"], cache[mbid]["high_level"]

    base_url = f"https://acousticbrainz.org/{mbid}/"

    def fetch_data(endpoint):
        while True:
            response = requests.get(base_url + endpoint)
            if response.status_code == 200:
                return response.json()
            elif response.status_code == 429:  # Rate limit exceeded
                retry_after = int(response.headers.get("Retry-After", 5))  # Default tunggu 5 detik
                print(f"Rate limit exceeded. Waiting {retry_after} seconds before retrying...")
                time.sleep(retry_after)
            else:
                return None

    # Request untuk low-level dan high-level data
    low_level = fetch_data("low-level")
    high_level = fetch_data("high-level")

    # Simpan ke cache
    cache[mbid] = {"low_level": low_level, "high_level": high_level}
    with open(cache_file, "w") as f:
        json.dump(cache, f)

    return low_level, high_level

# Loop untuk mengambil data setiap lagu

def process_mbid_data(input_file, output_file, start_row, end_row):
  data = []

  df = pd.read_csv(input_file, dtype=str)  # Pastikan membaca MBID sebagai string

  df['MBID'] = df['MBID'].astype(str).apply(clean_mbid)  # Bersihkan MBID

  selected_df = df.iloc[start_row-1:end_row]  # Pilih baris sesuai rentang

  for mbid in selected_df['MBID']:
      print(f"Fetching data for MBID: {mbid}")
      low_level, high_level = get_acousticbrainz_data(mbid)

      if low_level and high_level:
          # Ambil karakteristik yang relevan
          song_data = {
              "MBID": mbid,
              "Tempo": low_level.get("rhythm", {}).get("bpm", None),
              "Loudness": low_level.get("lowlevel", {}).get("average_loudness", None),

              "Danceability": high_level.get("highlevel", {}).get("danceability", {}).get("value", None),
              "Mood_Sad": high_level.get("highlevel", {}).get("mood_sad", {}).get("value", None),
              "Mood_Happy": high_level.get("highlevel", {}).get("mood_happy", {}).get("value", None),

              "Genre Dortmund": high_level.get("highlevel", {}).get("genre_dortmund", {}).get("value", None),
              "Genre Electronic": high_level.get("highlevel", {}).get("genre_electronic", {}).get("value", None),
              "Genre Rosamerica": high_level.get("highlevel", {}).get("genre_rosamerica", {}).get("value", None),
              "Genre Tzanetakis": high_level.get("highlevel", {}).get("genre_tzanetakis", {}).get("value", None),
          }
          data.append(song_data)
          print(f"Song Feature: {song_data}")
      else:
          print(f"Data not found for MBID: {mbid}")

      # Tambahkan delay untuk menghindari rate limiting
      time.sleep(2)

  result_df = pd.DataFrame(data)
  result_df.to_csv(output_file, index=False)  # Simpan hasil dalam format CSV
  print(f"Data telah disimpan ke {output_file}")

input_file = "file.csv"  # Nama file input yang berisi daftar MBID
output_file = "output_data.csv"  # Nama file output hasil
start_row = 1  # Baris mulai di CSV
end_row = 1115  # Baris akhir di CSV

process_mbid_data(input_file, output_file, start_row, end_row)
```
