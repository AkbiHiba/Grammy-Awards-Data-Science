```python
# Calculate percentage
import matplotlib.pyplot as plt

country_percent = grammys_only_artists['Country'].value_counts(normalize=True) * 100
country_percent = country_percent.sort_values(ascending=False)
# Match colors for existing countries in the data
bar_colors = [country_colors.get(country, 'lightgray') for country in country_percent.index]

plt.figure(figsize=(10, 6))
ax = country_percent.plot(kind='barh', color=bar_colors, edgecolor='black')

plt.title('Country Frequency in Grammys')
plt.ylabel('Country')
plt.xlabel('Percentage (%)')
ax.invert_yaxis()

# Add values at the end of bars
for i, v in enumerate(country_percent):
    ax.text(v + 0.5, i, f"{v:.1f}%", va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
country_percent_win = grammys_only_artists[grammys_only_artists['nb wins']>=1]['Country'].value_counts(normalize=True) * 100

# Plot
plt.figure(figsize=(12, 6))
ax = country_percent_win.plot(kind='barh', color='lightgreen',edgecolor='black')

plt.title('Country Frequency in Grammy Wins')
plt.xlabel('Country')
plt.ylabel('Percentage (%)')
ax.invert_yaxis()
# Add values on top of bars


for i, v in enumerate(country_percent_win):
    ax.text(v + 0.5,i, f"{v:.1f}%", va='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
country_percent_win = grammys_only_artists[grammys_only_artists['nb wins'] >= 1]['Country'].value_counts(normalize=True) * 100
country_percent_nom = grammys_only_artists['Country'].value_counts(normalize=True) * 100

# Combine into a single DataFrame
country_df = pd.concat([country_percent_nom, country_percent_win], axis=1)
country_df.columns = ['Nominated', 'Win']
country_df = country_df.fillna(0)  # Fill missing values with 0

# Plot grouped bar chart
ax = country_df.plot(kind='bar', figsize=(12, 6), edgecolor='black', color=['skyblue', 'lightcoral'])

plt.title('Country Frequency in Grammys')
plt.xlabel('Country')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=45)
plt.legend(title='Status')


# Add values on top of bars
for i, country in enumerate(country_df.index):
    for j, status in enumerate(['Nominated', 'Win']):
        val = country_df.loc[country, status]
        if val > 0:
            ax.text(i + (j - 0.5)*0.2, val + 0.5, f"{val:.1f}%", ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Get actual counts
win_counts = grammys_only_artists[grammys_only_artists['nb wins'] >= 1]['Country'].value_counts()
nom_counts = grammys_only_artists['Country'].value_counts()

# Compute win ratio per country
win_ratio = (win_counts / nom_counts) * 100

# Optional: Sort or select top countries
win_ratio = win_ratio.dropna().sort_values(ascending=False)
# Map colors based on country
bar_colors = [country_colors.get(country, 'lightgray') for country in win_ratio.index]

# Plot
plt.figure(figsize=(12, 6))
ax = win_ratio.plot(kind='barh', color=bar_colors, edgecolor='black')

plt.title('Win Ratio by Country (Wins / Nominations)')
plt.xlabel('Country')
plt.ylabel('Win Ratio (%)')
ax.invert_yaxis()

# Add percentage text above bars
for i, v in enumerate(win_ratio):
    ax.text( v + 0.5,i, f"{v:.1f}%", va='center', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
artist_wins = grammys_only_artists.groupby(['Country', 'Main Artist'])['nb wins'].sum().reset_index()

# For each country, get the artist with the most wins
top_artists_per_country = artist_wins.sort_values(['Country', 'nb wins'], ascending=[True, False]).groupby('Country').head(3)
# Optional: Sort by wins for readability
top_artists_per_country = top_artists_per_country.sort_values(by='Country', ascending=False)

print(top_artists_per_country)

artist_nom = grammys_only_artists.groupby(['Country', 'Main Artist'])['nb nominations'].sum().reset_index()

# For each country, get the artist with the most wins
top_artists_per_country = artist_nom.sort_values(['Country', 'nb nominations'], ascending=[True, False]).groupby('Country').head(3)
# Optional: Sort by wins for readability
top_artists_per_country = top_artists_per_country.sort_values(by='Country', ascending=False)

print(top_artists_per_country)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Merge the two summaries into one DataFrame
top_wins_per_country = artist_wins.sort_values(['Country', 'nb wins'], ascending=[True, False]).groupby('Country').head(5)
top_noms_per_country = artist_nom.sort_values(['Country', 'nb nominations'], ascending=[True, False]).groupby('Country').head(5)

top_combined = pd.merge(
    top_wins_per_country, top_noms_per_country,
    on=['Country', 'Main Artist'],
    how='outer'
).fillna(0)

top_combined = top_combined.astype({'nb wins': int, 'nb nominations': int})

# Plot: One subplot per country
unique_countries = top_combined['Country'].unique()
n_countries = len(unique_countries)

fig, axes = plt.subplots(
    nrows=(n_countries + 1) // 2, ncols=2,
    figsize=(16, 5 * ((n_countries + 1) // 2))
)

axes = axes.flatten()

for i, country in enumerate(unique_countries):
    ax = axes[i]
    data = top_combined[top_combined['Country'] == country]
    data_melted = data.melt(
        id_vars='Main Artist',
        value_vars=['nb wins', 'nb nominations'],
        var_name='Metric',
        value_name='Count'
    )

    sns.barplot(
        data=data_melted,
        x='Main Artist', y='Count', hue='Metric',
        palette={'nb wins': 'gold', 'nb nominations': 'skyblue'},
        ax=ax
    )

    ax.set_title(f"Top Artists in {country}")
    ax.set_ylabel("Count")
    ax.set_xlabel("")
    ax.tick_params(axis='x', rotation=30)

# Remove empty subplots if any
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
import numpy as np
import pandas as pd

# Filter to top countries
top_countries = grammys_data['Country'].value_counts().head(10).index
filtered_data = grammys_data[grammys_data['Country'].isin(top_countries)].copy()

# Define categories and assign a base Y for each
category_order = sorted(filtered_data['Category'].unique())
category_base = {cat: i * 10 for i, cat in enumerate(category_order)}  # Each category spaced by 10 units

# Map countries to offset within each category
country_offset = {country: j for j, country in enumerate(sorted(top_countries))}

# Compute composite Y position: base + offset + jitter
filtered_data['Category_Base'] = filtered_data['Category'].map(category_base)
filtered_data['Country_Offset'] = filtered_data['Country'].map(country_offset)
np.random.seed(42)
filtered_data['Y_Pos'] = filtered_data['Category_Base'] + filtered_data['Country_Offset'] + np.random.uniform(-0.4, 0.4, size=len(filtered_data))

# Set marker type
filtered_data['Marker'] = filtered_data['Win'].apply(lambda x: 'x' if x == 1 else 'circle')

# Plot
fig = px.scatter(
    filtered_data,
    x='Year',
    y='Y_Pos',
    color='Country',
    symbol='Marker',
    symbol_map={'circle': 'circle', 'x': 'x'},
    color_discrete_map={
        'United States': '#4C78A8',
        'United Kingdom': '#F3A72C',
        'Canada': '#6DBF5A',
        'Australia': '#bba5d4',
        'Uganda': '#F1C232',
        'Sweden': '#FF5F68',
        'The Carrebian': "gray",
        'Ireland': "gray"
    },
    hover_data=['Year', 'Country', 'Win', 'Main Artist', 'Cleaned Song or Album', 'Category'],
)

# Add dark edges to all points
fig.update_traces(
    marker=dict(size=12, line=dict(width=0.8, color='black')),
)

# Custom Y ticks: show Country name under each category block
tickvals = []
ticktext = []
for cat in category_order:
    base = category_base[cat]
    for country in sorted(top_countries):
        tickvals.append(base + country_offset[country])
        ticktext.append(f"{cat} - {country}")

fig.update_yaxes(
    tickvals=tickvals,
    ticktext=ticktext,
    title='Category - Country',
)

fig.update_layout(
    title='Nominations & Wins by Country and Category Over Time',
    legend_title='Country',
    height=800
)

fig.show()
```
