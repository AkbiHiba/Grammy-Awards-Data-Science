```python
def full_categorical_summary(df,columns):
    for col in columns:
        print(f"\n Summary for '{col}':")
        print(f"Unique values: {df[col].nunique()}")
        print(f"Most frequent value: {df[col].mode()[0] if not df[col].mode().empty else 'None'}")
        print("\nValue counts:")
        print(df[col].value_counts())
        print("\nPercentage distribution:")
        print(df[col].value_counts(normalize=True) * 100)
        print("\nMissing values:")
        print(df[col].isnull().sum())
        print(df[col].value_counts(normalize=True) * 100)
        missing_count = df[col].isnull().sum()
        missing_percentage = (missing_count / len(df)) * 100
        print(f"\nMissing values: {missing_count} ({missing_percentage:.2f}%)")
        print("-" * 50)

    return df.describe(include='all')
```

```python
import plotly.express as px
import pandas as pd

# Filter out unwanted rows
filtered_df = artists_df[
    ~((artists_df['nb nominations'] == 1) & (artists_df['nb wins'] == 1)) &
    (artists_df['nb wins'] != -1)
].copy()

# Define custom colors and assign a group for unknown artists
artist_colors = {
    'Taylor Swift': 'red',
    'Kanye West': 'black',
    'Billie Eilish': 'lightgreen',
    'Kendrick Lamar': 'blue',
    'U2': "darkkhaki",
    'Adele': "pink",
    'Jay-Z': 'gold',
    'Eminem': 'lightsteelblue',
    'Drake':'lavenderblush',
    'Beyonce':'thistle'
}
default_color = "lightgray"

# Group unknown artists into one category
filtered_df['Display Artist'] = filtered_df['Main Artist'].apply(
    lambda x: x if x in artist_colors else 'Other'
)

# Add "Other" to the color map
full_color_map = artist_colors.copy()
full_color_map['Other'] = default_color

# Plot
fig = px.scatter(
    filtered_df,
    x='nb nominations',
    y='nb wins',
    color='Display Artist',
    color_discrete_map=full_color_map,
    hover_name='Main Artist',
    title='Nominations vs Wins (Interactive)'
)

# Make dots more visible with darker edges
fig.update_traces(marker=dict(
    size=12,  # Increase size of dots
    line=dict(
        width=2,  # Thickness of border
        color='black'  # Dark color for the edge
    )
))

# Add red dotted line at y = 0
fig.add_shape(
    type="line",
    x0=filtered_df['nb nominations'].min(),
    y0=0,
    x1=filtered_df['nb nominations'].max(),
    y1=0,
    line=dict(color="red", width=2, dash="dot")
)

fig.update_layout(
    xaxis_title='Number of Nominations',
    yaxis_title='Number of Wins',
    height=600
)

fig.show()
```

```python
artists_in_charts= songs_in_charts_df.groupby('ArtistID').agg(
    Total_Chart_Weeks=('Nb of Weeks on Chart', 'sum'),
    Nb_Charted_Songs=('Nb of Weeks on Chart', lambda x: (x > 0).sum())
).reset_index()
artists_in_charts
```

```python
artists_charting= pd.merge(artists_in_charts, artists_df[['ArtistID', 'Main Artist','Gender', 'Country', 'State ID', 'nb wins', 'nb nominations', 'win rate']], on= 'ArtistID', how='left' )
```

```python
artists_charting #2267 rows Ã— 10 columns
```

```python
def categorize(row):
    if row['nb wins'] > 0:
        return 'Winner'
    elif row['nb wins'] == 0 and row['nb nominations']>0:
        return 'Nominated (Non-Winner)'
    else:
        return 'Non-Nominated'

def categorize_bynominated(row):
    if row['nb nominations'] > 0:
        return 'Nominated'
    else:
        return 'Not Nominated'

artists_charting['Category'] = artists_charting.apply(categorize, axis=1)
artists_charting['Is Nominated'] = artists_charting.apply(categorize_bynominated, axis=1)
```

```python
from scipy.stats import mannwhitneyu

winners = artists_charting[artists_charting['Category'] == 'Winner']['Nb_Charted_Songs']
non_winners = artists_charting[artists_charting['Category'] == 'Nominated (Non-Winner)']['Nb_Charted_Songs']

# Mann-Whitney U Test
stat, p = mannwhitneyu(winners, non_winners, alternative='two-sided')
print(f"Mann-Whitney U test on NB CHARTED SONGS (win vs non win ) p-value: {p}")


winners = artists_charting[artists_charting['Is Nominated'] == 'Nominated']['Nb_Charted_Songs']
non_winners = artists_charting[artists_charting['Is Nominated'] == 'Not Nominated']['Nb_Charted_Songs']

# Mann-Whitney U Test
stat, p = mannwhitneyu(winners, non_winners, alternative='two-sided')
print(f"Mann-Whitney U test on NB CHARTED SONGS (nominated vs non nominated ) p-value: {p}")




winners = artists_charting[artists_charting['Category'] == 'Winner']['Total_Chart_Weeks']
non_winners = artists_charting[artists_charting['Category'] == 'Nominated (Non-Winner)']['Total_Chart_Weeks']

# Mann-Whitney U Test
stat, p = mannwhitneyu(winners, non_winners, alternative='two-sided')
print(f"Mann-Whitney U test on NB TOTAL CHARTED WEEKS (win vs non win ) p-value: {p}")


winners = artists_charting[artists_charting['Is Nominated'] == 'Nominated']['Total_Chart_Weeks']
non_winners = artists_charting[artists_charting['Is Nominated'] == 'Not Nominated']['Total_Chart_Weeks']

# Mann-Whitney U Test
stat, p = mannwhitneyu(winners, non_winners, alternative='two-sided')
print(f"Mann-Whitney U test on NB TOTAL CHARTED WEEKS (nominated vs non nominated ) p-value: {p}")
```
