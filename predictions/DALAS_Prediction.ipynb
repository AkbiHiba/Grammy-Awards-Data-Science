```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score
import numpy as np
import pandas as pd

def cross_val_pred(clf, X, y):
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    reports = []
    accuracies = []

    # For ROC
    tprs = []
    aucs = []
    mean_fpr = np.linspace(0, 1, 100)

    plt.figure(figsize=(8,6))

    i = 0
    for train_idx, val_idx in skf.split(X, y):
        i += 1
        X_tr, X_val = X[train_idx], X[val_idx]
        y_tr, y_val = y[train_idx], y[val_idx]

        clf.fit(X_tr, y_tr)
        y_pred = clf.predict(X_val)
        y_proba = clf.predict_proba(X_val)[:, 1]  # prob for positive class

        acc = accuracy_score(y_val, y_pred)
        accuracies.append(acc)

        report = classification_report(y_val, y_pred, output_dict=True)
        print(f'Fold {i} ------------')
        print(pd.DataFrame(report).T)
        reports.append(report)

        # ROC curve for this fold
        fpr, tpr, _ = roc_curve(y_val, y_proba)
        roc_auc = roc_auc_score(y_val, y_proba)
        aucs.append(roc_auc)

        # Interpolate TPR for mean FPR points
        interp_tpr = np.interp(mean_fpr, fpr, tpr)
        interp_tpr[0] = 0.0
        tprs.append(interp_tpr)

        # Plot fold ROC
        plt.plot(fpr, tpr, lw=1, alpha=0.3, label=f'ROC fold {i} (AUC = {roc_auc:.2f})')

    # Plot mean ROC curve
    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = np.mean(aucs)
    std_auc = np.std(aucs)

    plt.plot(mean_fpr, mean_tpr, color='b',
             label=f'Mean ROC (AUC = {mean_auc:.2f} ± {std_auc:.2f})',
             lw=2, alpha=0.8)

    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', lw=2)
    plt.xlim([0, 1])
    plt.ylim([0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve - Cross-Validated')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()

    # Aggregate accuracy
    print(f"Average accuracy over folds: {np.mean(accuracies):.4f}")

    # Aggregate classification report (macro avg)
    def average_reports(reports):
        classes = [c for c in reports[0].keys() if c not in ('accuracy', 'macro avg', 'weighted avg')]
        avg_report = {}
        for cls in classes + ['macro avg', 'weighted avg']:
            avg_report[cls] = {}
            for metric in reports[0][cls].keys():
                avg_report[cls][metric] = np.mean([r[cls][metric] for r in reports])
        return avg_report

    avg_report = average_reports(reports)
    print('**********************************')
    df_report = pd.DataFrame(avg_report).T
    print(df_report)

```

```python
from sklearn.decomposition import PCA

def run_pca(df, numerical_features, n_components=None):
    numerical_data = df[numerical_features].copy()

    # Impute missing values with mean for numerical columns
    numerical_imputer = SimpleImputer(strategy="mean")
    numerical_data = numerical_imputer.fit_transform(numerical_data)

    # Scale the numerical features
    scaler = StandardScaler()
    numerical_scaled = scaler.fit_transform(numerical_data)

    # Apply PCA
    pca = PCA(n_components=n_components)  # set n_components to None or int
    numerical_pca = pca.fit_transform(numerical_scaled)

    print(f"Explained variance by PCA components: {pca.explained_variance_ratio_}")
    print(f"Cumulative explained variance: {pca.explained_variance_ratio_.cumsum()}")

    return pca, numerical_pca
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_pca_variance(pca):
    explained_variance = pca.explained_variance_ratio_
    cumulative_variance = np.cumsum(explained_variance)

    plt.figure(figsize=(12, 5))

    # Plot cumulative explained variance
    plt.subplot(1, 2, 1)
    plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.title('Cumulative Explained Variance by PCA Components')
    plt.grid(True)

    # Plot variance decrease (elbow/scree plot)
    plt.subplot(1, 2, 2)
    plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
    plt.xlabel('Number of Components')
    plt.ylabel('Explained Variance Ratio')
    plt.title('Explained Variance per Component (Scree Plot)')
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
def plot_pca_variance_with_elbow(pca):
    explained_variance = pca.explained_variance_ratio_
    cumulative_variance = np.cumsum(explained_variance)

    plt.figure(figsize=(12, 5))

    # Plot cumulative explained variance
    plt.subplot(1, 2, 1)
    plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.title('Cumulative Explained Variance by PCA Components')
    plt.grid(True)

    # Plot variance decrease (elbow/scree plot)
    plt.subplot(1, 2, 2)
    plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
    plt.xlabel('Number of Components')
    plt.ylabel('Explained Variance Ratio')
    plt.title('Explained Variance per Component (Scree Plot)')
    plt.grid(True)

    # Attempt to automatically find the elbow point (a simple heuristic)
    diffs = np.diff(explained_variance)
    elbow_index = np.argmax(np.diff(diffs)) + 1  # Find the point of maximum change in the rate of decrease

    # Draw a vertical line at the estimated elbow
    plt.axvline(x=elbow_index + 1, color='r', linestyle='--', label=f'Estimated Elbow at {elbow_index + 1} Components')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_correlation_circle(pca, features, scale=1.0):
    pcs = pca.components_[:2]  # first two PCs
    plt.figure(figsize=(8, 8))
    plt.axhline(0, color='grey', lw=1)
    plt.axvline(0, color='grey', lw=1)

    for i, feature in enumerate(features):
        # Scale component loadings for better visualization
        x = pcs[0, i] * scale
        y = pcs[1, i] * scale
        plt.arrow(0, 0, x, y, head_width=0.03, head_length=0.03, fc='red', ec='red')
        plt.text(x * 1.15, y * 1.15, feature, color='black', ha='center', va='center')

    circle = plt.Circle((0, 0), 1, facecolor='none', edgecolor='b', linestyle='--')
    plt.gca().add_artist(circle)

    plt.xlim(-1.1, 1.1)
    plt.ylim(-1.1, 1.1)
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.title('PCA Correlation Circle')
    plt.grid()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def pca_biplot(pca, X_pca, features, labels=None, scale=1.0):
    pcs = pca.components_[:2]
    xs = X_pca[:, 0]
    ys = X_pca[:, 1]

    plt.figure(figsize=(10, 8))

    # Scatter plot of samples projected on PC1 and PC2
    plt.scatter(xs, ys, alpha=0.6, edgecolors='k', label='Samples')

    # If you have labels (like categories), you can plot them with different colors:
    if labels is not None:
        for label in np.unique(labels):
            idx = labels == label
            plt.scatter(xs[idx], ys[idx], label=label, alpha=0.7)

    # Draw arrows for features (variables)
    for i, feature in enumerate(features):
        x = pcs[0, i] * scale
        y = pcs[1, i] * scale
        plt.arrow(0, 0, x, y, head_width=0.03, head_length=0.03, fc='red', ec='red')
        plt.text(x * 1.15, y * 1.15, feature, color='black', ha='center', va='center')

    # Draw unit circle for correlation reference
    circle = plt.Circle((0, 0), 1, facecolor='none', edgecolor='b', linestyle='--')
    plt.gca().add_artist(circle)

    plt.xlabel('PC1 (%.2f%% explained variance)' % (pca.explained_variance_ratio_[0] * 100))
    plt.ylabel('PC2 (%.2f%% explained variance)' % (pca.explained_variance_ratio_[1] * 100))
    plt.title('PCA Biplot')
    plt.grid()
    plt.legend()
    plt.axhline(0, color='grey', lw=1)
    plt.axvline(0, color='grey', lw=1)
    plt.show()
```

```python
from sklearn.decomposition import PCA

def preprocess_numerical_with_pca(df, numerical_features, n_components=None):
    numerical_data = df[numerical_features].copy()

    # Impute missing values
    numerical_imputer = SimpleImputer(strategy="mean")
    numerical_data = numerical_imputer.fit_transform(numerical_data)

    # Scale numerical features
    scaler = StandardScaler()
    numerical_scaled = scaler.fit_transform(numerical_data)

    # Apply PCA
    pca = PCA(n_components=n_components)
    numerical_pca = pca.fit_transform(numerical_scaled)

    return numerical_pca, scaler, pca
```
